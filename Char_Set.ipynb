```python
import string
def only_in_target(path):
    symbol_sets = []
    symbols = []
    
    content = open(path, 'r').readlines()

    lang_count = LANG_CHAR_COUNT[lang]
    file_count = {}

    for line in content:
        line = line.split('	')
        src = line[0].strip()
        trg = line[1].strip()
        punc = string.punctuation
        
        for c in trg:
            if c in punc and c not in src:
                if c not in symbols:
                    symbols.append(c)
                    symbol_sets.append({
                        "symbol": c,
                        "src": src,
                        "trg": trg,
                        "sym_count": 1
                    })
                else:
                    i = next((x for x in symbol_sets if x["symbol"] == c), None)
                    i["sym_count"] = i["sym_count"] + 1
    return symbol_sets
```

```python
def count_char2(lang, path):
#     print(path)
    content = open(path, 'r').readlines()

    lang_count = LANG_CHAR_COUNT[lang]
    file_count = {}
    
    for line in content:
        word = line.split('	')[1]
        word = word.strip()
#         word = word.lower()
#         print("word:", word)
#         print("set:", set(word))
        
        for c in set(word):
            if c in lang_count:
                lang_count[c] = int(lang_count[c]) + word.count(c)
            else:
                lang_count[c] = word.count(c)
            
            if c in file_count:
                file_count[c] = int(file_count[c]) + word.count(c)
            else:
                file_count[c] = word.count(c)
                
    return file_count
```

```python
import os

folder_path = "src-data/news2018/news2018/data/"
target_suffix = "_strn.v1.txt.trg"
data_dir = os.listdir(folder_path)
LANG = {}

for lang_dir in data_dir:
    if lang_dir == ".DS_Store": # stupid MAC
        continue
    for f in os.listdir(os.path.join(folder_path, lang_dir)):
        if f.endswith(target_suffix):
            file_path = os.path.join(folder_path, lang_dir, f)
            file = open(file_path, 'r').read()
            uni_chars = set(file)
            target_lang = lang_dir[2:4] # last 2 chars, "EnCh => Ch"
            print("Language =>", target_lang)
            print("Unique Chars =>", len(uni_chars))           
            if target_lang in LANG:
                if len(uni_chars) > len(LANG[target_lang]):
                    LANG[target_lang] = uni_chars
            else:
                LANG[target_lang] = uni_chars
```
